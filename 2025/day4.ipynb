```python
from AOC_utils import get_day
import numpy as np

input_data = get_day(day=4, year=2025)

# print out first few lines to get a feel for the data
print(len(input_data))
input_data[:7]
```

```python
example = '''
..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.
'''.split('\n')[1:-1]
```

```python
def solve_day(data):
    part1 = 0; part2 = 0

    data = np.array([[y for y in x] for x in data]) # 2d array

    # padd the array with extra dots
    data = np.pad(data, pad_width=1, mode='constant', constant_values='.')

    for i in range(1, data.shape[0]-1):
        for j in range(1, data.shape[1]-1):
            cell = data[i,j]
            if cell == '@':
                neighbors = [data[i-1,j-1], data[i-1,j], data[i-1,j+1],
                             data[i,j-1],               data[i,j+1],
                             data[i+1,j-1], data[i+1,j], data[i+1,j+1]]
                rolls_of_paper = neighbors.count('.')
                
                if rolls_of_paper > 4:
                    part1 += 1

    while True:
        count = 0
        for i in range(1, data.shape[0]-1):
            for j in range(1, data.shape[1]-1):
                cell = data[i,j]
                if cell == '@':
                    neighbors = [data[i-1,j-1], data[i-1,j], data[i-1,j+1],
                                data[i,j-1],               data[i,j+1],
                                data[i+1,j-1], data[i+1,j], data[i+1,j+1]]
                    rolls_of_paper = neighbors.count('.')
                    
                    if rolls_of_paper > 4:
                        data[i,j] = '.'
                        count += 1
                        part2 += 1
        if count == 0:
            break

    print("part 1:", part1)
    print("part 2:", part2)

solve_day(example)
```

```python
solve_day(input_data)
```
