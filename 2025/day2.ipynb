```python
example = '''
11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124
'''.split('\n')[1:-1]
```

```python
def check_number(num):
    for l in range(1, (len(num)//2)+1):
        subsect = num[:l]
        if num[l:] == subsect:
            return True
    return False

def check_number_p2(num):
    arr = np.array(list(num))
    for l in range(1, (len(num)//2)+1):
        if len(num) % l == 0:
            arr = arr.reshape(-1, l)
            if np.all(arr[0] == arr):
                return True
    return False

def solve_day(data):
    part1 = 0
    part2 = 0

    data = [x.strip().split(',') for x in data][0]
    for pair in data:
        first, second = pair.split('-')
        for num in range(int(first), int(second)+1):
            num_str = str(num)
            if check_number(num_str):
                part1 += num
            if check_number_p2(num_str):
                part2 += num

    print("part 1:", part1)
    print("part 2:", part2)

solve_day(example)
```

```python
solve_day(input_data)
```
