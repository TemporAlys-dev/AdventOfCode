```python
example = '''
987654321111111
811111111111119
234234234234278
818181911112111
'''.split('\n')[1:-1]
```

```python
def find_max_joltages(banks, n_batteries):
    total = 0
    for bank in banks:
        best_indexs = np.arange(n_batteries)
        for i in range(n_batteries):
            best_indexs[i] += np.argmax(bank[best_indexs[i]:len(bank)-((n_batteries-1) - i)])
            if i != n_batteries-1: best_indexs[i+1] = best_indexs[i] + 1
        total += int(''.join([bank[x] for x in best_indexs]))
    return total

def solve_day(data):
    data = np.array([[y for y in x] for x in data])
            
    print("part 1:", find_max_joltages(data, 2))
    print("part 2:", find_max_joltages(data, 12))

solve_day(example)
```
