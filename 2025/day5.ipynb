```python
example = '''
3-5
10-14
16-20
12-18

1
5
8
11
17
32
'''.split('\n')[1:-1]
```

```python
def solve_day(data):
    part1 = 0; part2 = 0

    fresh_ranges = []
    ingredients = []
    for line in data:
        if '-' in line:
            fresh_ranges.append([int(x) for x in line.split('-')])
        elif line != '':
            ingredients.append(int(line))

    while True:
        new_ranges = []
        for range1 in fresh_ranges:
            for range2 in fresh_ranges:
                if not (range1[1] < range2[0] - 1 or range1[0] > range2[1] + 1) and not range1 == range2:
                    new_range = [min(range1[0], range2[0]), max(range1[1], range2[1])]
                    if new_range not in new_ranges:
                        new_ranges.append(new_range)
                    break
            else:
                if range1 not in new_ranges:
                    new_ranges.append(range1)
        if len(new_ranges) == len(fresh_ranges):
            break
        fresh_ranges = new_ranges

    for range_ in fresh_ranges:
        part2 += (range_[1] - range_[0]) + 1
    
    # run part 1 after merging ranges
    for ingredient in ingredients:
        for range_ in fresh_ranges:
            if ingredient >= range_[0] and ingredient <= range_[1]:
                part1 += 1
                break

    print("part 1:", part1)
    print("part 2:", part2)

solve_day(example)
```

```python
solve_day(input_data)
```
