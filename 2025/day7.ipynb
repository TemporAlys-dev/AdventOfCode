```python
from AOC_utils import get_day
import numpy as np

input_data = get_day(day=7, year=2025)

# print out first few lines to get a feel for the data
print(len(input_data))
input_data[:7]
```

```python
example = '''
.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............
'''.split('\n')[1:-1]
```

```python
def solve_day(data):
    part1 = 0; part2 = 0

    data = np.array([list(x) for x in data])[::2] # 2d array
    data = np.pad(data, ((0,1),(0,0)), mode='constant', constant_values='.')

    beam_counts_array = np.zeros(data.shape, dtype=np.int64)

    Sloc = np.where(data == 'S')

    beam_counts_array[Sloc[0], Sloc[1]] = 1

    beam_locs = np.array([Sloc[0], Sloc[1]]).T
    
    for i in range(0, data.shape[0]-1):

        beam_counts = beam_counts_array[beam_locs[:,0], beam_locs[:,1]]
        below_locs = beam_locs + np.array([1, 0])[None,:]

        data_slice = data[below_locs[:,0], below_locs[:,1]]
        mask = data_slice == '.'
        beam_counts_array[below_locs[mask,0], below_locs[mask,1]] += beam_counts[mask]

        mask = data_slice == '^'
        part1 += np.sum(mask)

        new_beamsL = below_locs[mask,:] + np.array([0, -1])[None,:]
        new_beamsR = below_locs[mask,:] + np.array([0, 1])[None,:]

        beam_counts_array[new_beamsL[:,0], new_beamsL[:,1]] += beam_counts[mask]
        beam_counts_array[new_beamsR[:,0], new_beamsR[:,1]] += beam_counts[mask]

        beam_locs = np.argwhere(beam_counts_array[i+1] > 0)
        beam_locs = np.concatenate([np.full(len(beam_locs), i+1)[:,None], beam_locs], axis=1)

    part2 += np.sum(beam_counts_array[-2])

    print("part 1:", part1)
    print("part 2:", part2)

solve_day(example)
```

```python
solve_day(input_data)
```
