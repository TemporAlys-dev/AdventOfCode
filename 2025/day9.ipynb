```python
example = '''
7,1
11,1
11,7
9,7
9,5
2,5
2,3
7,3
'''.split('\n')[1:-1]
```

```python
import shapely
```

```python
def solve_day(data):
    part1 = 0; part2 = 0

    data = np.array([x.split(',') for x in data], dtype=np.int64)

    all_squares = (np.abs(data[...,None,:] - data[None,...,:])+1).prod(axis=-1)

    part1 = np.max(all_squares)

    ngon = shapely.geometry.Polygon(data)
    
    best_area = 0
    for i in range(len(data)):
        corner1 = data[i]
        for j in range(i+1, len(data)):
            corner2 = data[j]

            corner3 = np.array([corner1[0], corner2[1]])
            corner4 = np.array([corner2[0], corner1[1]])
            
            box_shape = shapely.geometry.Polygon([corner1, corner3, corner2, corner4])

            if ngon.contains(box_shape):
                box_area = (np.abs(corner2 - corner1)+1).prod()
                if box_area > best_area:
                    best_area = box_area
                    
    part2 = best_area

    print("part 1:", part1)
    print("part 2:", part2)

solve_day(example)
```

```python
solve_day(input_data)
```
