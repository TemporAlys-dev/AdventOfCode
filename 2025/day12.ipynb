```python
from AOC_utils import get_day
import numpy as np

input_data = get_day(day=12, year=2025)

# print out first few lines to get a feel for the data
print(len(input_data))
input_data[:7]
```

```python
example = '''
0:
###
##.
##.

1:
###
##.
.##

2:
.##
###
##.

3:
##.
###
##.

4:
###
#..
###

5:
###
.#.
###

4x4: 0 0 0 0 2 0
12x5: 1 0 1 0 2 2
12x5: 1 0 1 0 3 2
'''.split('\n')[1:-1]
```

```python
def solve_day(data):
    part1 = 0

    shapes = []
    trees = []
    current_shape = []
    for line in data:
        if line == '':
            if current_shape:
                shapes.append(np.array(current_shape, dtype=int))
                current_shape = []
        elif 'x' not in line:
            if ':' in line:
                continue
            else:
                current_shape.append([c == '#' for c in line])
        else:
            size, counts = line.split(':')
            w,h = map(int, size.split('x'))
            counts = list(map(int, counts.strip().split()))
            trees.append((w,h,counts))
    
    shape_totals = [np.sum(s) for s in shapes]

    pruned_trees = []
    for w, h, counts in trees:
        tree_size = w * h
        total_presents_size = np.array(counts) * (np.array(shape_totals) * 1.2) # increase to account for inefficiency
        if np.sum(total_presents_size) <= tree_size:
            pruned_trees.append((w,h,counts))
    trees = pruned_trees
    part1 = len(trees)
    
    print("part 1:", part1)

solve_day(example)
```

```python
solve_day(input_data)
```
