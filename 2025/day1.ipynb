```python
example = '''
L68
L30
R48
L5
R60
L55
L1
L99
R14
L82
'''.split('\n')[1:-1]
```

```python
def solve_day(data):
    part1 = 0; part2 = 0

    direction = [x[0] for x in data]
    distance = [int(x[1:]) for x in data]
    
    dial_at = 50
    for turn, dist in zip(direction, distance):
        dial_at += dist * (1 if turn == 'R' else -1)

        if abs(dial_at // 100) > 0:
            part2 += abs(dial_at // 100)

        dial_at = dial_at % 100

        if dial_at == 0:
            part1 += 1

    print("part 1:", part1)
    print("part 2:", part2)

solve_day(example)
```

```python
def solve_day(data):
    part1 = 0; part2 = 0

    direction = [x[0] for x in data]
    distance = [int(x[1:]) for x in data]
    
    dial_at = 50
    for turn, dist in zip(direction, distance):
        for _ in range(dist):
            dial_at += 1 if turn == 'R' else -1
            if dial_at > 99:
                dial_at = 0
            elif dial_at < 0:
                dial_at = 99
                
            if dial_at == 0:
                part2 += 1

        dial_at = dial_at % 100
        if dial_at == 0:
            part1 += 1

    print("part 1:", part1)
    print("part 2:", part2)

solve_day(example)
```

```python
solve_day(input_data)
```
